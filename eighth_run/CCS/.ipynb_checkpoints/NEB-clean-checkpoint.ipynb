```python
from ase import io
from ase.neb import NEB
from ase.optimize import MDMin

# Read initial and final states:
initial = io.read('opt.traj')
final = io.read('final_opt.traj')

# Make a band consisting of 5 images:
nimage = 5
images = [init]
images += [init.copy() for i in range(nimage)]
images += [final]
neb = NEB(images)

# Interpolate linearly the potisions of the three middle images:
neb.interpolate()

# Set calculators:
for image in images[1:(nimage+1)]:
    CCS_calc = CCS(CCS_params=CCS_params)
    image.calc = CCS_calc
    
# Optimize:
optimizer = MDMin(neb, trajectory='A2B.traj')
optimizer.run(fmax=0.05)
```
